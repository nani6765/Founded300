```python
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import UnexpectedAlertPresentException
from selenium.webdriver.common.by import By
```

```python
while True:
    try: 
        btnBox = driver.find_element_by_class_name('btn_box3_txt')
        btnBox.click()
    except UnexpectedAlertPresentException:
        break
```

```python
from bs4 import BeautifulSoup

html = driver.page_source 
soup = BeautifulSoup(html, 'html.parser') # BeautifulSoup사용하기
divs = soup.select('a > div.startup_info > div.info-box')
```

```python
import math
import locale
```

```python
name= []
value= []

for div in divs:
    subject = div.find("div", {"class" : "subject"})
    pureSubject = subject.text.replace(" ","").replace("\t", "")
    
    mogip = div.find("div", {"class": "dal"}).find("div", {"class", "dal2"})
    pureMogip = mogip.text.replace(" ","").replace("\t", "").replace("%", "").replace(",", "")
    calMogip = 500000 * int(float(pureMogip))
    finalMogip = "{:,}".format(math.floor(calMogip))
    
    name.append(pureSubject)
    value.append(finalMogip)
```

```python
from pandas import Series, DataFrame
import pandas as pd
```

```python
raw_data = {'이름': name,
            '금액': value,}

data = pd.DataFrame(raw_data)
data.to_csv('result.csv', encoding='utf-8-sig', mode='w')
```
